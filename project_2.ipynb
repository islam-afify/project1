I will  drop some columns  I do not use to be efficient for data analysis

I will replace zero  values in columns budget ,  revenue , runtime with means in this each column

```python
#replacing zero  values in columns budget ,  revenue , runtime  with  means in this each column
statmov[ 'budget' ]=statmov[ 'budget' ].replace(0,statmov[ 'budget' ].mean())
statmov['revenue'] =statmov['revenue'].replace(0,statmov['revenue'].mean())
statmov[ 'runtime' ]=statmov[ 'runtime' ].replace(0,statmov[ 'runtime' ].mean())
```

```python
statmov.describe()
```

To make generes column more normalize 

```python
statmov['genres'] = statmov['genres'].apply(lambda x: x.split('|')[0])
```

I will create new function "print_mini_max_values" to avoid reapting code while getting the highest and lowest 

values ...for example the lowest and highest values of budget of movies 

```python
def print_mini_max_values(my_parameter):

    index_lowest = statmov[my_parameter].idxmin()


    index_highest = statmov[my_parameter].idxmax()

    low_value = pd.DataFrame(statmov.loc[index_lowest])
    high_value = pd.DataFrame(statmov.loc[index_highest])

#show the movie with lowest and highest values 
 
    print("The movie with lowest"+ " " + my_parameter + "  " + "  " , statmov['original_title'][index_lowest])

    print("The movie with highest" +  " " + my_parameter + "  " + "  " , statmov['original_title'][index_highest])

    return pd.concat([low_value,high_value],axis = 1)
```
